```python
import matplotlib.pyplot as plt
import numpy as np
from math import sin,sqrt
```

```python
x = np.arange(-5,6,0.2)
y = np.arange(-5,6,0.2)
```

```python
v = []
for i in x:
    v.append(sin(sqrt((i**2)*2)))
print(v)
z = np.array(v)
```

```python
ax = plt.axes(projection = '3d')
ax.scatter3D(x, y,z)
```
